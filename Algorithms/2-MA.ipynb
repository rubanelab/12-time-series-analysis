```python
whole_data = pd.read_csv("dataset\Tatacoffee13_21.csv")
```

```python
whole_data
```

```python
stk_data = whole_data[
    (whole_data["Date"] >= "2021-07-01") & (whole_data["Date"] <= "2022-01-28")
]

print(stk_data)
```

```python
stk_data = stk_data[["Open", "High", "Low", "Close"]]
```

```python
stk_data
```

```python
column = "Close"
```

```python
scaler = MinMaxScaler()
data1 = scaler.fit_transform(stk_data[[column]])

print("Len:", data1.shape)
```

```python
len(data1)
```

```python
import sklearn
print(sklearn.__version__)
```

```python
training_size = round(len(data1 ) * 0.80)
print(training_size)

X_train = data1[:training_size]
X_test = data1[training_size:]

print("X_train length:", X_train.shape)
print("X_test length:", X_test.shape)

y_train = data1[:training_size]
y_test = data1[training_size:]

print("y_train length:", y_train.shape)
print("y_test length:", y_test.shape)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
trends = ['n','t','c','ct']
orders = [(0, 0, 1), (0, 0, 2)]

for trend in trends:
    # print(td)
    # trendslist.append(td)
    model = ARIMA(X_train, order=(0, 0, 10), trend=trend)
    model_fit = model.fit()
    
    # make prediction
    start = len(X_train)
    end = len(X_train) + len(y_test) - 1

    y_pred = model_fit.predict(start=start, end=end)
    # y_pred= model_fit.predict(len(X_train), len(data1) - 1)
    
    #print(y_pred)
    #mse=mean_squared_error(y_test,y_pred,squared=False)
    
    print("Trend={}".format(trend))
    # print(y_test, y_pred)
    rmsemape(y_test, y_pred)
    print(" -------------------------------- ")
```
